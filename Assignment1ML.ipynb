```python
import numpy as np
```

```python
# Q1

# (a)
arr = np.array([1, 2, 3, 6, 4, 5])
reversed_arr = arr[::-1]
print("Q1(a):", reversed_arr)
```

```python
# (b)
array1 = np.array([[1, 2, 3], [2, 4, 5], [1, 2, 3]])
flattened_1 = array1.flatten()
flattened_2 = array1.ravel()
print("Q1(b) Method 1:", flattened_1)
print("Q1(b) Method 2:", flattened_2)
```

```python
#(c)
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])
comparison = np.array_equal(arr1, arr2)
element_wise = arr1 == arr2
print("Q1(c) Arrays equal:", comparison)
print("Q1(c) Element-wise:", element_wise)
```

```python
# (d)
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])

unique_x, counts_x = np.unique(x, return_counts=True)
most_frequent_x = unique_x[np.argmax(counts_x)]
indices_x = np.where(x == most_frequent_x)[0]
print("Q1(d)i Most frequent in x:", most_frequent_x, "Indices:", indices_x)

unique_y, counts_y = np.unique(y, return_counts=True)
most_frequent_y = unique_y[np.argmax(counts_y)]
indices_y = np.where(y == most_frequent_y)[0]
print("Q1(d)ii Most frequent in y:", most_frequent_y, "Indices:", indices_y)
```

```python
# (e)
gfg = np.matrix('4, 1, 9; 12, 3, 1; 4, 5, 6')
sum_all = np.sum(gfg)
sum_rows = np.sum(gfg, axis=1)
sum_cols = np.sum(gfg, axis=0)
print("Q1(e)i Sum of all elements:", sum_all)
print("Q1(e)ii Sum row-wise:", sum_rows.flatten())
print("Q1(e)iii Sum column-wise:", sum_cols.flatten())
```

```python
# (f)
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
diag_sum = np.trace(n_array)
eigenvalues, eigenvectors = np.linalg.eig(n_array)
inverse_matrix = np.linalg.inv(n_array)
determinant = np.linalg.det(n_array)
print("Q1(f)i Sum of diagonal:", diag_sum)
print("Q1(f)ii Eigenvalues:", eigenvalues)
print("Q1(f)iii Eigenvectors:", eigenvectors)
print("Q1(f)iv Inverse:", inverse_matrix)
print("Q1(f)v Determinant:", determinant)
```

```python
# (g)
p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])
mult_1 = np.dot(p1, q1)
cov_1 = np.cov(p1.flatten(), q1.flatten())
print("Q1(g)i Multiplication:", mult_1)
print("Q1(g)i Covariance:", cov_1)

p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])
mult_2 = np.dot(p2, q2)
cov_2 = np.cov(p2.flatten(), q2.flatten())
print("Q1(g)ii Multiplication:", mult_2)
print("Q1(g)ii Covariance:", cov_2)
```

```python
# (h)
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
inner_prod = np.inner(x, y)
outer_prod = np.outer(x, y)
cartesian_prod = np.cross(x, y)
print("Q1(h) Inner product:", inner_prod)
print("Q1(h) Outer product:", outer_prod)
print("Q1(h) Cartesian product:", cartesian_prod)
```

```python
# Q2

# (a)
array = np.array([[1, -2, 3],[-4, 5, -6]])
abs_array = np.abs(array)
print("Q2(a)i Absolute values:", abs_array)

flattened = array.flatten()
perc_25_flat = np.percentile(flattened, 25)
perc_50_flat = np.percentile(flattened, 50)
perc_75_flat = np.percentile(flattened, 75)
print("Q2(a)ii Percentiles (flattened) - 25th:", perc_25_flat, "50th:", perc_50_flat, "75th:", perc_75_flat)

perc_25_col = np.percentile(array, 25, axis=0)
perc_50_col = np.percentile(array, 50, axis=0)
perc_75_col = np.percentile(array, 75, axis=0)
print("Q2(a)ii Percentiles (columns) - 25th:", perc_25_col, "50th:", perc_50_col, "75th:", perc_75_col)

perc_25_row = np.percentile(array, 25, axis=1)
perc_50_row = np.percentile(array, 50, axis=1)
perc_75_row = np.percentile(array, 75, axis=1)
print("Q2(a)ii Percentiles (rows) - 25th:", perc_25_row, "50th:", perc_50_row, "75th:", perc_75_row)

mean_flat = np.mean(flattened)
median_flat = np.median(flattened)
std_flat = np.std(flattened)
print("Q2(a)iii Flattened - Mean:", mean_flat, "Median:", median_flat, "Std:", std_flat)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)
print("Q2(a)iii Columns - Mean:", mean_col, "Median:", median_col, "Std:", std_col)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)
print("Q2(a)iii Rows - Mean:", mean_row, "Median:", median_row, "Std:", std_row)
```

```python
# (b)
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
floor_vals = np.floor(a)
ceil_vals = np.ceil(a)
trunc_vals = np.trunc(a)
round_vals = np.round(a)
print("Q2(b) Floor:", floor_vals)
print("Q2(b) Ceiling:", ceil_vals)
print("Q2(b) Truncated:", trunc_vals)
print("Q2(b) Rounded:", round_vals)
```

```python
# Q3:

# (a)
array = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_array = np.sort(array)
sorted_indices = np.argsort(array)
smallest_4 = np.partition(array, 3)[:4]
largest_5 = np.partition(array, -5)[-5:]
print("Q3(a)i Sorted array:", sorted_array)
print("Q3(a)ii Sorted indices:", sorted_indices)
print("Q3(a)iii 4 smallest:", np.sort(smallest_4))
print("Q3(a)iv 5 largest:", np.sort(largest_5))
```
